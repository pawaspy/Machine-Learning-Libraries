# Matplotlib example workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt

# 1.Prepare data

x=[1,2,5,9]
y=[5,1,9,2]

# 2.Setup plot
fig,ax=plt.subplots(figsize=(10,10))

# 3.Plot data
ax.plot(x,y)

# 4.Set plot
ax.set(title="Sample plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5.Save figure
fig.savefig("numpy-images/sample-plot.png")
```

## Two options for subplots

```python
# Subplots option 1
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,
                                      ncols=2,
                                      figsize=(10,5))
# Plot to each different
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig,ax=plt.subplots(nrows=2,
                   ncols=2,
                   figsize=(10,5))
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

### Which one should we use?
*When plotting something, okay to use the pyplot method
*When plotting something more advanced, use the OO method


```python
over_50=heart_disease[heart_disease["age"]>50]
```

```python
over_50.head()
```

```python
over_50.plot(kind="scatter",
           x="age",
           y="chol",
           c="target");
```

```python
# OO method
fig,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax);
```

```python
fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"]);
ax.set(title="Heart disease and Cholestrol level",
       xlabel="Age",
       ylabel="Cholestrol")
ax.legend(*scatter.legend_elements(),title="Target");
ax.axhline(over_50["chol"].mean(),
           linestyle='--');
```

```python
# Create plot (2)
fig,(ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,10),
                           sharex=True)
# Add data to ax0
scatter=ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);
# Customize ax0
ax0.set(title="Heart Disease and Cholestrol level",
        ylabel="Cholestrol");
# Add a legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target");
# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--");

# Add data to ax1
scatter=ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"]);
# Customize ax1
ax1.set(title="Heart disease and Maximum heart rate",
        xlabel="Age",
        ylabel="Maximum heart rate");
# Add legend
ax1.legend(*scatter.legend_elements(),title="Target");
# Add meanline
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--");

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")
```

```python
# Customize our plot with set()
ax=df.plot(kind='bar')
ax.set(title="Random number bar graph from dataframe",
       xlabel="Row number",
       ylabel="Random number")
ax.legend().set_visible(True)
```

```python
plt.style.use("seaborn-whitegrid")
fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"],
                   cmap="winter");
ax.set(title="Heart disease and Cholestrol level",
       xlabel="Age",
       ylabel="Cholestrol")
ax.legend(*scatter.legend_elements(),title="Target");
ax.axhline(over_50["chol"].mean(),
           linestyle='--');
```

```python
# Customising the lines
# Create plot (2)
fig,(ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,10),
                           sharex=True)
# Add data to ax0
scatter=ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="winter");
# Customize ax0
ax0.set(title="Heart Disease and Cholestrol level",
        ylabel="Cholestrol");
ax0.set_xlim([50,80])
# Add a legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target");
# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--");

# Add data to ax1
scatter=ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"],
                    cmap="winter");
# Customize ax1
ax1.set(title="Heart disease and Maximum heart rate",
        xlabel="Age",
        ylabel="Maximum heart rate");
ax1.set_xlim([50,80])
ax1.set_ylim([6,200])
# Add legend
ax1.legend(*scatter.legend_elements(),title="Target");
# Add meanline
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--");

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
